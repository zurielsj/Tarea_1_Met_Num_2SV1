3. Escribir una función que reciba como argumentos el número de horas y el costo por hora dejarlo
como argumento predefinido en 150 la función debe de regresar el pago del empleado en un
día, en una semana (5 días) y en un mes (20 días).

```
def sueldo_por_tiempo(horas,sueldo):
  sueld=horas*sueldo
  print("El sueldo es",sueld,"por dia")
  print("El sueldo es",sueld*5,"por semana")
  print("El sueldo es",sueld*20,"por mes")
a=float(input("Horas trabajadas: "))
b=150
sueldo_por_tiempo(a,b)  
```

opción elegida por el usuario se desarrolle la operación, observe que en la división el denominador no puede ser cero, en caso que sea así, debe de entrar en un bucle while que esté leyendo
el denominador de nuevo hasta que éste sea distinto de cero.


```
print(               "~~~~Calculadora~~~~") 
print(               "Menu de operaciones") 
print("1.-Suma,  2.-Resta,  3.-Multiplicacion, 4.-Division")
resultado=""
while resultado!="no": 
  Menu=['1) Suma','2) Resta','3) Multiplicacion','4) Division']
  print("Elija una opcion del menu")
  Menu=int(input())
  while Menu>=5: 
    print("Elija una opcion del menu")
    Menu= int(input()) 

  if Menu==1: 
    x= float(input("Introduzca el primer valor: "))
    y= float(input("Introduzca el segundo valor: "))
    print("La suma de", x, "y de", y, "es: ", x+y) 

  elif Menu==2: 
    z= float(input("Introduzca el primer valor: "))
    t= float(input("Introduzca el segundo valor: "))
    print("La resta de", z, "y de", t, "es: ", z-t) 

  elif Menu==3: 
    a= float(input("Introduzca el primer valor: "))
    b= float(input("Introduzca el segundo valor: "))
    print("La multiplicacion de", a, "y de", b, "es: ", a*b) 

  elif Menu==4: 
    k= float(input("Introduzca el primer valor: "))
    l= float(input("Introduzca el segundo valor: "))
    while k==0:
      print("Digite un numero distinto de 0")
      l=float(input())
    print("La division de", k, "y de", l, "es: ", k/l)

  print("¿Desea realizar otra operacion?")
  resultado=input()

print("Fin del programa")
```
